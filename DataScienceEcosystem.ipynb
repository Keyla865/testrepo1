### Some of the popular languages that Data Scientists use are:

1. **Python**: One of the most used languages, with libraries such as pandas, NumPy, scikit-learn, TensorFlow, among others.
2. **R**: Wide use in statistical analysis and data visualization, with packages such as ggplot2 and dplyr.
3. **SQL**: Fundamental language for database management and querying.
4. **Julia**: Popular for her high performance in numerical calculations.
5. **Scala**: Used in conjunction with Apache Spark for real-time analysis and large volumes of data.
6. **Java**: Used in the development of large systems and distributed applications.
7. **MATLAB**: Widely used in academic and scientific environments for numerical calculations and data analysis.
8. **SAS**: Software and language specialized in statistical analysis, especially in business environments.

# Some of the commonly used libraries used by Data Scientists include:

1. **pandas**: Python library for structured data manipulation and analysis.
2. **NumPy**: Support for arrays and high-level math operations in Python.
3. **Matplotlib**: Library for data visualization in 2D graphs in Python.
4. **Seaborn**: Matplotlib extension, focused on creating attractive and easy-to-interpret statistical graphs.
5. **scikit-learn**: Set of tools for machine learning in Python, with classification, regression and clustering algorithms.
6. **TensorFlow**: Open source library for deep learning developed by Google.
7. **Keras**: High-level API for creating and training deep neural networks, running on top of TensorFlow and Theano.
8. **PyTorch**: Deep learning library developed by Facebook, popular for its ease of use and flexibility.
9. **Statsmodels**: Python library to perform statistical tests and estimate statistical models.
10. **SciPy**: NumPy extension that provides functions for integration, optimization, linear algebra, etc.
